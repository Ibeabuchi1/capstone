```python
#seperate the features and the target
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#split the data to training and testing set
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=None)
```

```python
#create an instance and hyper parameters tuning
classifier = RandomForestClassifier()
parameters = {'n_estimators': range(10, 50, 10),
             'max_depth': range(1, 12, 2),
             'min_samples_leaf': range(1,7),
             'min_samples_split': range(2,9,2)}
grid_search = GridSearchCV(classifier, parameters, cv=5, n_jobs=1, verbose=1)
grid_search.fit(X_train, y_train)
grid_search.best_params_
```

```python
#using the best estimators from the parameters
best_estimator = grid_search.best_estimator_
best_estimator.score(X_test,y_test)
```

```python
#creating a dataframe to view the estimators
feature_importance = best_estimator.feature_importances_
feature_importance_df = pd.DataFrame({'features': list(X_train),
                                     'feature-importance': feature_importance})
```

```python
feature_importance_df = feature_importance_df.sort_values('feature-importance', ascending=False)
feature_importance_df
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
x_pred = best_estimator.predict(X_test)
```

```python
print(confusion_matrix(x_pred, y_test))
```

```python
print(accuracy_score(x_pred, y_test))
```

```python
#plot the result in a pie chart
feature_importance_df.plot.pie(
explode= [0.1]*len(X_train.columns),
labels = feature_importance_df.features,
y = 'feature-importance',
autopct = '%1.1f%%',
legend = False,
shadow = True,
figsize = (8,8))
```
